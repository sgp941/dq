# Visualising Job Outcomes

In this project we are analysing data on job outcomes released by the American Community Survey, which conducts surveys and aggregates the data. The cleaned dataset we will be using is released on the FiveThirtyEight [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Some of the main columns in the dataset are:

- **Rank** - Rank by median earnings (the dataset is ordered by this column).
- **Major_code** - Major code.
- **Major** - Major description.
- **Major_category** - Category of major.
- **Total** - Total number of people with major.
- **Sample_size** - Sample size (unweighted) of full-time.
- **Men** - Male graduates.
- **Women** - Female graduates.
- **ShareWomen** - Women as share of total.
- **Employed** - Number employed.
- **Median** - Median salary of full-time, year-round workers.
- **Low_wage_jobs** - Number in low-wage service jobs.
- **Full_time** - Number employed 35 hours or more.
- **Part_time** - Number employed less than 35 hours.

## Scatter

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(5,10))
plt.grid()
```

With an increasing sample size, we see the median salary tending towards smaller range between 30k-50k

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size', figsize=(5,10))
plt.grid()
```

With higher sample size we see the unemployment rate averaging out at between 5-7%

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median', figsize=(5,10))
plt.grid()
```

With higher rates of full-time employment, we reach a similar median salary as with the greater sample size, at between 30k-50k.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment_rate', figsize=(5,10))
plt.grid()
```

There does not seem to be any discernable relationship between the proportion of women on the course, and the unemployment rate. Suggesting that the gender distribution of the course does not have an effect on employability.

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(5,10))
plt.grid()
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(5,10))
plt.grid()
```

The courses with larger numbers of men tend to have a wider variety of median salaries than those with larger numbers of women. The bottom of the range remains similar across both groups, near to 30k, however the men achieve a much higher range, up to 60k. This is almost 10k higher than courses with similar numbers of women. 

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs. Median', figsize=(5,10))
plt.grid()
```

Comparing the median salary with the proportion of women supports the analysis from the previous set of graphs, showing a clear negative correlation. That is, the higher the proportion of women in the course, the lower the median salary. This is in contrast to the effect of proportion of women with employability. 

```python
recent_grads.plot(x='Total', y='Median', kind='scatter', title='Total vs. Median', figsize=(5,10))
plt.grid()
```

Here we actually see that the more popular the course, the lower the median salary. This is likely due to the smaller proportion of graduates achieving the high-end of salaries are now not as visible and do not have as much of an effect on a larger 

## Hist

```python
recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
```

Most of the course data seems to be from smaller sample sizes, which could indicate smaller courses, but most probably actually indicates lower involvement in the surveys.

```python
recent_grads['Median'].hist(bins=15, range=(20000,80000))
```

From this, we can see the most common median salary range is around 35k. We can also see a somewhat normal distribution around this median. 

```python
recent_grads['Employed'].hist(bins=15, range=(0,40000))
```

Here we clearly see that higher numbers of employed graduates is less common, most likely due to fewer courses containing the larger numbers of students. 

```python
recent_grads['Full_time'].hist(bins=10, range=(0,30000))
```

```python
recent_grads['ShareWomen'].hist(bins=10, range=(0,1))
```

The proportion of women across courses largely seems to follow slightly skewed normal distribution. Slightly higher than 50% of majors are predominantly female.

```python
recent_grads['Unemployment_rate'].hist(bins=10, range=(0,.15))
```

Unemployment rate seems to follow a normal distribution between 0 and 15%, with a median range between 5-10%.
