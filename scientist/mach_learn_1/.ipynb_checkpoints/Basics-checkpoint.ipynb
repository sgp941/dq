# Predicting Car Prices

This project will attempt to predict a car's market value by using data on it's attributes. The data set we will be working with contains information on various cars. For each car we have information about the technical aspects of the vehicle such as the motor's displacement, the weight of the car, the miles per gallon, how fast the car accelerates, and more. 

More information on this data set, made available by UC Irvine, is available [here](https://archive.ics.uci.edu/ml/datasets/automobile).

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', names=cols)
cars.head()
```

```python
# select only the columns with continuous values from - https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names
continuous_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cont_cars = cars[continuous_cols]
cont_cars.head()
```

## Data cleaning

```python
cont_cars = cont_cars.replace('?', np.nan)
cont_cars = cont_cars.astype('float')

cont_cars.isnull().sum()
```

There are three options for theee missing values:

1. Replace the missing values using the average values from that column.
2. Drop the rows entirely (especially if other columns in those rows have missing values).
3. Drop the column entirely.

Because `price` is the column we want to predict, let's remove any rows with missing `price` values.

```python
cont_cars = cont_cars.dropna(subset=['price'])
cont_cars.isnull().sum()
```

Replace missing values in other columns using column means.

```python
cont_cars = cont_cars.fillna(cont_cars.mean())
cont_cars.isnull().sum()
```

```python
# normalize all columnns to range from 0 to 1 except the target column.
price_col = cont_cars['price']
norm_cars = (cont_cars - cont_cars.min())/(cont_cars.max() - cont_cars.min())
norm_cars['price'] = price_col
```

## Univariate model

With some basic cleaning out of the way, let's start with some univariate k-nearest neighbors models.

```python
def knn_train_test(train_col, target, df):
    # Divide number of rows in half and round.
    last_train_row = int(len(df) / 2)
    
    # training/testing split 70/30
    split = round(df.shape[0] * 0.7)
    train_df = df.iloc[:split]
    test_df = df.iloc[split:]
    
    # Fit a KNN model using default k value.
    knn = KNeighborsRegressor()
    knn.fit(train_df[[train_col]], train_df[target])
    
    # Make predictions using model.
    predicted_labels = knn.predict(test_df[[train_col]])

    # Calculate and return RMSE.
    mse = mean_squared_error(test_df[target], predicted_labels)
    rmse = np.sqrt(mse)
    return rmse
```

```python
rmse_results = {}
train_cols = norm_cars.columns.drop('price')

# for each column (minus `price`), train a model, return RMSE value
# and add to the dictionary `rmse_results`.
for col in train_cols:
    rmse_val = knn_train_test(col, 'price', norm_cars)
    rmse_results[col] = rmse_val

# convert to Series for manipulation
rmse_results_series = pd.Series(rmse_results)
rmse_results_series.sort_values()
```

```python
def knn_train_test(train_col, target, df, k):        
    # Randomize order of rows in data frame.
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    # training/testing split 70/30
    split = round(rand_df.shape[0] * 0.7)
    train_df = rand_df.iloc[:split]
    test_df = rand_df.iloc[split:]
    
    # Fit a KNN model using default k value.
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(train_df[[train_col]], train_df[target])
    
    # Make predictions using model.
    predicted_labels = knn.predict(test_df[[train_col]])

    # Calculate and return RMSE.
    mse = mean_squared_error(test_df[target], predicted_labels)
    rmse = np.sqrt(mse)
    return rmse
```

```python
train_cols = norm_cars.columns.drop('price')

rmse_results = {}
k_values = [1, 3, 5, 7, 9]
for col in train_cols:
    rmse_k = []
    for k in k_values:
        rmse_val = knn_train_test(col, 'price', norm_cars, k)
        rmse_k.append(rmse_val)
    rmse_results[col] = rmse_k

rmse = pd.DataFrame(rmse_results)
rmse['k_val'] = k_values
rmse = rmse.set_index('k_val')
rmse
```

```python
rmse.plot()
plt.legend(loc='right', bbox_to_anchor=(1.5, .5))
```

## Multivariate model

Let's modify the knn_train_test() function we wrote in the last step to work with multiple columns.

```python
def knn_train_test(train_cols, target, df, k=5):        
    # Randomize order of rows in data frame.
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    # training/testing split 70/30
    split = round(rand_df.shape[0] * 0.7)
    train_df = rand_df.iloc[:split]
    test_df = rand_df.iloc[split:]
    
    # Fit a KNN model using default k value.
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(train_df[train_cols], train_df[target])
    
    # Make predictions using model.
    predicted_labels = knn.predict(test_df[train_cols])

    # Calculate and return RMSE.
    mse = mean_squared_error(test_df[target], predicted_labels)
    rmse = np.sqrt(mse)
    return rmse
```

```python
# compute average RMSE across different `k` values for each feature.
feature_avg_rmse = rmse.mean().sort_values()
feature_avg_rmse
```

```python
rmse_results = {}
for i in range(2,6):
    train_cols = feature_avg_rmse[:i].index.tolist()
    rmse_val = knn_train_test(train_cols, 'price', norm_cars)
    rmse_results[i] = rmse_val

rmse_results = pd.Series(rmse_results)
rmse_results = rmse_results.rename('RMSE')
rmse_results
```

## Hyperparameter tuning

Let's now optimize the model that performed the best in the previous step.

```python
train_cols = []
for i in [3, 4, 5]:
    train_cols.append(feature_avg_rmse[:i].index.tolist())
    
rmse_results = {}
k_values = range(1, 26)
for i, cols in enumerate(train_cols):
    rmse_k = []
    for k in k_values:
        rmse_val = knn_train_test(cols, 'price', norm_cars, k)
        rmse_k.append(rmse_val)
    rmse_results[i] = rmse_k

rmse = pd.DataFrame(rmse_results)
rmse['k_val'] = k_values
rmse = rmse.set_index('k_val')
rmse.columns = ['3 features', '4 features', '5 features']
rmse
```

```python
rmse.plot()
plt.ylabel('RMSE')
plt.xlabel('KNN K-Value')
plt.grid(b=True)
```

The best choice of k depends upon the data; generally, larger values of k reduces effect of the noise on the classification, but make boundaries between classes less distinct. This explains why we see larger error rates with higher K-values. 

The trade-off is visible for all three models somewhat in the middle where the RMSE flattens for a short period while the K-value is increasing. From this exploration we would choose the following K-values:

- 3 features -> k=10
- 4 features -> k=16
- 5 features -> k=17
